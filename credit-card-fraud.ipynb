```python
def prep_data(dataset: pd.DataFrame) -> (np.ndarray, np.ndarray):
    X = dataset.iloc[:, 2:30].values
    y = dataset.Class.values
    return X, y

def plot_data(X: np.ndarray, y: np.ndarray):
    plt.scatter(X[y == 0, 0], X[y == 0, 1], label="Class #0", alpha=0.5, linewidth=0.10)
    plt.scatter(X[y == 1, 0], X[y == 1, 1], label="Class #1", alpha=0.5, linewidth=0.10, c='r') 
    plt.legend()
    return plt.show()

X, y = prep_data(dataset)
plot_data(X, y)
```
